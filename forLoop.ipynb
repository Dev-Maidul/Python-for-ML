### For Loop

```python
for i in range(1,11):
    for j in range(1,11,2):
        print(i,j,"Hello World")
```

### While Loop

```python
count=1;
while count<11:
    print(count,"Hello world")
    count=count+1
```

### Continue and Break

```python
sum=0
for i in range(1,11):
    print(i,"is processing")
    if(i%2==0):
        continue
    print(i,"is added to the sum")
    sum+=i
    print(sum)
```

### Break

```python
accuracy=95
while True:
    if accuracy==101:
        break
    print(accuracy)
    accuracy+=1
```
