```python
n=int(input())
yes=0;
no=0;
for _ in range(n):
    vote=input()
    if(vote=="YES"):
        yes=yes+1
    elif(vote=="NO"):
        no=no+1
if(yes>=no):
    print("ACCEPT")
else:
    print("REJECT")
        
```

### AI Mood Detector

```python
message=input()
if("happy" in message) or ("joy" in message) or ("smile" in message):
    print("Happy Mood")
elif("sad" in message) or ("sad" in message) or ("angry" in message):
    print("Sad Mood")
else:
    print("Neutral Mood")
```

### AI Brightness Analyzer

```python
inp=input()
arr=[]
sum=0;
for num in inp.split():   
    arr.append(int(num))
    sum+=int(num)

lenth=(len(arr))
avg=int(sum/lenth)
if(avg<85):
    print("Dark Image")
elif(avg>=170):
    print("Bright Image")
else:
    print("Normal Image")
```

### AI brightness analizer

```python
inp=input()
arr=inp.split()
count_A=arr.count('A')
count_B=arr.count('B')
total=len(arr)
if count_A > 0.7 * total or count_B > 0.7 * total:
    print("Biased Model")
else:
    print("Fair Model")
```
