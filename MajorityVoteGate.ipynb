```python
n=int(input())
yes=0;
no=0;
for _ in range(n):
    vote=input()
    if(vote=="YES"):
        yes=yes+1
    elif(vote=="NO"):
        no=no+1
if(yes>=no):
    print("ACCEPT")
else:
    print("REJECT")
        
```

### AI Mood Detector

```python
message=input()
if("happy" in message) or ("joy" in message) or ("smile" in message):
    print("Happy Mood")
elif("sad" in message) or ("sad" in message) or ("angry" in message):
    print("Sad Mood")
else:
    print("Neutral Mood")
```
