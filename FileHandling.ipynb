Write File


```python
# Create a new file if file not exist and write there
with open("./sample_data/test.txt","w") as file:
  file.write("Hello student\n")
  file.write("How is going class?\n")
```

Append File

```python
with open("./sample_data/test.txt","a") as file:
  file.writelines("What you learn about file Handling?\n")
  file.writelines("Do you have any issues about course?\n")
```

```python
text=["Hello\n","hi\n","whatsup\n"]
with open("./sample_data/test.txt","a") as file:
  for tst in text:
    file.writelines(tst)
    file.writelines("\n")
```

File Pointer

```python
with open("./sample_data/test.txt","r") as file:
  # tell method mean where is my cursor/pointer
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  # seek method means we can move cursor backward and forward
  print(file.seek(10))
  print(file.read(5))
  print(file.tell())
```


Problem 1  Create and Write to a File


Write a function create_file(filename, content) that creates a new text file with the given filename and writes the content into it. Call it using positional arguments.

📥 Sample Input:
"notes.txt", "Hello, this is my first file."
📤 Sample Output (notes.txt will contain):
Hello, this is my first file.


```python
filename=input()
content=input()
def create_file(filename, content):
  with open(f"./sample_data/{filename}","w") as file:
    file.write(content)

create_file(filename,content)
```

Write a function read_lines(filename) that reads a text file and prints each line with its line number.

📥 Sample Input (file: data.txt):
Apple
Banana
Cherry
📤 Sample Output:
1: Apple
2: Banana
3: Cherry


```python
def read_lines(filename):
  with open(f"./sample_data/{filename}","r") as file:
    for i,line in enumerate(file,start=1):
      print(f"{i}: {line.strip()}")

read_lines("data.txt")
```
