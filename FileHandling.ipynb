File Read

```python
with open("./sample_data/sample.txt","r") as file:
  for line in file:
    # print(line,end="")
    l=line.strip()
    print(l)
```

Write File


```python
# Create a new file if file not exist and write there
with open("./sample_data/test.txt","w") as file:
  file.write("Hello student\n")
  file.write("How is going class?\n")
```

Append File

```python
with open("./sample_data/test.txt","a") as file:
  file.writelines("What you learn about file Handling?\n")
  file.writelines("Do you have any issues about course?\n")
```

```python
text=["Hello\n","hi\n","whatsup\n"]
with open("./sample_data/test.txt","a") as file:
  for tst in text:
    file.writelines(tst)
    file.writelines("\n")
```

File Pointer

```python
with open("./sample_data/test.txt","r") as file:
  # tell method mean where is my cursor/pointer
  print(file.tell())
  print(file.read(5))
  print(file.tell())
  # seek method means we can move cursor backward and forward
  print(file.seek(10))
  print(file.read(5))
  print(file.tell())
```

Exception Handling

```python
n=int(input())
try:
  a=10/0
except Exception as e:
  print(e)
else:
  print(10/n)
finally:
  print("Program stopped")
```
