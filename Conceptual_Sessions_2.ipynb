Serialization and Deserialization

```python
a = {
    'name': 'rakib',
    'age': 20,
}
import json
# Serialization
with open("sample.txt","w") as file:
  json.dump(a,file)
# Deserialization
with open("sample.txt","r") as f:
  print(json.load(f))
```

```python
L = [1, 2,3,4,5]
import json
with open("sample2.txt","w") as file:
  json.dump(L,file)
with open("sample2.txt","r") as f:
  print(json.load(f))
```
