### Codeforces problem K. Max and Min

```python
inp=input();
numbers=inp.split();
a=int(numbers[0]);
b=int(numbers[1]);
c=int(numbers[2]);
min=a;
max=a;
if(b<min):
    min=b;
if(c<min):
    min=c;

if(b>max):
    max=b;
if(c>max):
    max=c;
print(min,max)
```

### C. Even, Odd, Positive and Negative

```python
n=int(input());
inp=input();
numbers=inp.split();
even=0; odd=0;positive=0; negetive=0;
for i in range(n):
    x=int(numbers[i])
    if(x>0):
        positive+=1;
    elif (x<0):
        negetive+=1;
    if(x%2==0):
        even+=1;
    else:
        odd+=1;
    
print("Even:",even)
print("Odd:",odd)
print("Positive:",positive)
print("Negative:",negetive)

```
