### Dictionary Comprehension

```python
square={x:x**2 for x in range(10)}
# print(square)

## Even number square
even_square={y:y**2 for y in range (21) if y%2==0}
print(even_square)
words = ["apple", "banana", "cat"]
lenOfWord={word:len(word) for word in words}
print(lenOfWord)

```
